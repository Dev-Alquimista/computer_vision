#### OpenCv VideoCapture
1. VideoCapture 개요 
    * OpenCV의 VideoCapture class는 동영상을 개별 Frame으로 읽어들이는 기능을 제공 
    * VideoWriter는 VideoCaputre로 읽어들인 개별 Frame을 동영상 파일로 write수행  
    * VideoCapture 객체는 생성인자로 video 파일 위치를 받아서 생성 
    * VideoCapture 객체는 입력 Video 파일의 다양한 속성을 가져오는 것이 가능 
    * VideoCapture 객체의 read()는 마지막 Frame까지 차례로 Frame를 읽음 
2. VideoWriter 개요 
    * VideoWriter 객체는 write할 동영상의 파일 위치, Encoding codec유형, write fps 수치, frame 크기를 생성인자로 입력 받아 이들 값에 따른 동영상 write 수행 
    * VideoWriter는 write시 특정 포맷으로 동영상을 Encoding 가능

```python
import os 
import sys 
import cv2 
import matplotlib.pyplot as plt 
import time 
```

```python
VIDEO_IN_PATH = "/content/data/Night_Day_Chase.mp4"
VIDEO_OUT_PATH = "/content/data/Night_Day_Chase_out.mp4"
```

```python
cap = cv2.VideoCapture(VIDEO_IN_PATH)
codec = cv2.VideoWriter_fourcc(*"XVID")

video_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
video_fps = cap.get(cv2.CAP_PROP_FPS)

video_writer = cv2.VideoWriter(VIDEO_OUT_PATH, codec, video_fps, video_size)
frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"total frame count {frame_cnt}, \nFPS : {round(video_fps)}, \nFrame Size {video_size}")
```

```python
green_color  = (0, 255, 0)
red_color = (0, 0, 255)

start = time.time()
index =0 

while True : 
    hasFrame, img_frame = cap.read()

    if not hasFrame:
        print("There are no more frames to process")
        break
    index += 1
    print(f"frame : {index} completed")

    cv2.rectangle(img_frame, (300, 100, 800, 400), color = green_color, thickness=2) 
    caption = "frame:{}".format(index)
    cv2.putText(img_frame, caption, (300, 95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)

    video_writer.write(img_frame)

print(f"write time:{round(time.time()-start,3)}")
video_writer.release()
cap.release()
```
