#### Mean Average Precision 
1. Confidence 에 따른 Precision 과 Recall의 변화 
    * Example01 
        - Bird = 5, True Box=1 All Precision Box = 1
        - Precision = 1/1 , Recall = 1/5
    * Example02 
        - Bird = 5, True Box=2 All Precision Box = 2
        - Precision = 2/2 , Recall = 2/5
    * Example03 
        - Bird = 5, True Box=2 All Precision Box = 4 
        - Precision = 2/4 , Recall = 2/5

2. Mean Average Precision
    * Average Precision 은 한 개의 Object에 대한 성능 수치 
    * MAP는 여러 Object들의 Average Precision을 평균한 값 
    * PASCAL VOC는 IOU를 0.5로 고정했지만 COCO Challenge는 IOU 값을 다양한 범위로 설정하여 예측 성공 기준을 정하고 있다.
